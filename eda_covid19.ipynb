## Introduccion:

La siguiente base de Datos contiene informacion de los casos positivos de Covid 19 en el departamento del Atlantico entre el 20 de Marzo de 2020 hasta el 17 de Enero de 2024 reportado por los Datos Abiertos del Instituto Nacional de Salud en Colombia. El archivo contiene un total de 419061 datos y 23 diferentes variables que contienen datos importantes como fechas, datos demograficos, causa de contagio, diagnostico final, entre otros.

Se creará una nueva variable que permita calcular el tiempo de recuperación de los pacientes recuperados y el tiempo transcurrido hasta el fallecimiento en los casos de muerte. Esto facilitará comprender mejor el comportamiento y la evolución de la enfermedad en la población estudiada.

Adicionalmente con estos datos podremos identificar :
-  Cuáles son los casos más comunes de contagio de COVID-19 en el departamento del Atlántico.
- Analizar en qué tipo de población se concentraron la mayor cantidad de muertes. 
- Observar los tiempos en los que se registraron mas casos positivos de Covid 19.



## Descripcion de Variables:

**fecha reporte web** : Fecha en que se diligencio el caso en esta web.

**ID de caso** : Numero del caso positivo de covid 19.

**Fecha de Notificacion** : Fecha en que se reporto el resultado positivo de covid 19 del paciente.

**Nombre municipio** : Nombre del municipio donde el contagiado por covid 19 se realizo la prueba.

**Edad** : Edad del contagiado.

**Unidad de medida de edad** : Unidad de medida de edad del contagiado ( años : 1 , meses : 2 , dias : 3)

**Tipo de contagio**: Forma en que fue contagiado por covid 19.

**Ubicacion del caso**: Lugar donde transcurrio el contagiado por covid 19.

**Codigo ISO del país**: Codigo ISO del pais de origen del paciente.

**Recuperado**: Estado final del paciente.

**Fecha de inicio de sintomas** : Dia en el que tuvo los primeros sintomas el paciente.

**Fecha de muerte** : Dia en el que murio el paciente.

**Fecha de diagnostico** : Dia en el que se recibio el resultado positivo por Covid 19.

**Fecha de recuperacion** : Dia en el que se recupero el paciente del covid 19.

**Tipo de recuperacion**  : Forma en que se confirma el paciente como recuperado.

**Pertenencia Etnica** :  Etnia a la que pertenece el contagiado por Covid 19.(Indígena :1 , ROM: 2 , Raizal: 3, Palenquero :4 , Negro: 5 , Otro: 6).

**Nombre del grupo Etnico**: Nombre del grupo etnico que pertenece el paciente( en caso haga parte de uno).





```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns       
import numpy as np

df = pd.read_csv("C:/Users/Juanes Garcia Gomez/Documents/EDA#1/Casos_positivos_de_COVID-19_en_atlantico._20260221 (2).csv", sep=";")

df.head()

```

```python
df = df.dropna(how="all")
df.shape
```

## Creacion variable Dias de recuperacion

Tomamos la diferencia de dias entre la fecha de primeros sintomas y fecha de recuperacion

```python
df["Fecha de inicio de sintomas"] = pd.to_datetime(df['Fecha de inicio de sintomas'], dayfirst=True)
df["Fecha de recuperacion"] = pd.to_datetime(df["Fecha de recuperacion"], dayfirst=True)

df['Dias de recuperacion'] = (
    df['Fecha de recuperacion'] - df['Fecha de inicio de sintomas']
).dt.days
df["Dias de recuperacion"].describe()   
```

## Creacion Variable Dias hasta fallecimiento
Esta variable fue creada para calcular el número de días transcurridos desde la fecha de inicio de síntomas hasta la fecha de fallecimiento, con el fin de analizar el tiempo de evolución del covid 19 en los casos graves.


```python
df["Fecha de muerte"] = pd.to_datetime(df["Fecha de muerte"], dayfirst=True)
df["dias hasta fallecimiento"] = (df["Fecha de muerte"] - df["Fecha de inicio de sintomas"]).dt.days
df["dias hasta fallecimiento"].describe()
```

## Revision N/A

```python
resumen_na = pd.DataFrame({
    'Total NA': df.isna().sum(),
    'PORCENTAJE NA': df.isna().mean() * 100
}).sort_values(by='PORCENTAJE NA', ascending=False)

resumen_na
```

Teniendo en cuenta la base de datos y los datos N/A, los datos N/A en la variable "Nombre del Pais" son de Colombia por lo que se van a reemplazar por "Colombia. Por otra parte, por la poca utilidad de la variable Codigo ISO del país y su gran porcentaje de N/A, se va a borrar la columna.

```python
df["Nombre del Pais"] = df["Nombre del Pais"].fillna("Colombia")
```

```python
df = df.drop(columns=["Codigo ISO del país"])
```

Adicionalmente, al analizar los valores faltantes en la variable “Recuperado”, se observa que estos corresponden a registros que presentan una fecha de muerte, lo que indica que se trata de personas fallecidas.

```python
df["Recuperado"] = df["Recuperado"].fillna("Fallecido")
```

## Conversion Variable edad
La variable "Unidad de medida de edad" indica si la edad del paciente está expresada en años, meses o días. En este caso se va a unificar la información y se realizará la conversión de la edad de todos los pacientes a años.

```python
df["Edad_años"] = df["Edad"]  # base: años

df.loc[df["Unidad de medida de edad"] == 2, "Edad_años"] = (
    df.loc[df["Unidad de medida de edad"] == 2, "Edad"] / 12
)

df.loc[df["Unidad de medida de edad"] == 3, "Edad_años"] = (
    df.loc[df["Unidad de medida de edad"] == 3, "Edad"] / 365
)
```

```python
df["Fecha de notificacion"] = pd.to_datetime(df["Fecha de notificacion"])

tabla_casos = (
    df.groupby([
        df["Fecha de notificacion"].dt.year,
        df["Fecha de notificacion"].dt.month
    ])
    .size()
    .unstack(fill_value=0)
)

tabla_casos.columns = [
    "Enero","Febrero","Marzo","Abril","Mayo","Junio",
    "Julio","Agosto","Septiembre","Octubre","Noviembre","Diciembre"
]

tabla_casos
```

Por lo observado en la grafica y en la tabla, hubo un pico de casos positivos durante Marzo y Abril 2021 en el que se obtuvieron 53,350 y 71,715 casos positivos de Covid 19 en el Atlantico. Tambien se puede observar que en el año 2023 se redujieron considerablemente los casos. 

### Casos por Municipio:

```python
casos_municipio = (
    df.groupby("Nombre municipio")
    .size()
    .sort_values(ascending=True)
)

plt.figure(figsize=(8,8))
plt.barh(casos_municipio.index, casos_municipio.values)
plt.title("Casos positivos de COVID-19 por municipio en Atlántico")
plt.xlabel("Número de casos")
plt.ylabel("Municipio")
plt.show()
```

A partir del grafico de Barras, se puede observar que la gran mayoria de los casos positvos de covid 19 se concentraron en Barranquilla ,seguido por Soledad y Sabanalarga.

### Tipo de Contagio:

```python
plt.figure(figsize=(10, 6))
plt.pie(
    df["Tipo de contagio"].value_counts(),
    labels=df["Tipo de contagio"].value_counts().index,
    autopct="%1.3f%%",   
    startangle=90
)
plt.title("Distribución de tipos de contagio en casos positivos de COVID-19 en Atlántico")
plt.show()
```

Con base en los resultados, el tipo de contagio mas comun es el comunitario con un 68%, es decir, cuando no se sabe la fuente exacta del contagio. La segunda forma de contagio mas frecuente es la "relacionada" con un 31%, en donde se sabe que hubo contacto directo con otra persona ya con el virus. Por ultimo, la menos comun es la "importada", y trata de los casos en donde se contagio en el extranjero y la vino despues al pais.

## Recuperacion:



```python
df["Recuperado"] = df["Recuperado"].fillna("Fallecido")
Caso_recuperado = (
    df["Recuperado"]
    .value_counts()
    .loc[["Recuperado", "Fallecido"]]
)
plt.figure(figsize=(6,4))
plt.pie(
    Caso_recuperado.values,
    autopct="%1.1f%%",
    startangle=90
)
plt.legend(Caso_recuperado.index, loc="upper right")



plt.title("Distribución de casos recuperados y fallecidos de COVID-19 en Atlántico")
plt.show()
```

Segun el grafico circular, se puede observar que existe una tasa de mortalidad del 3% en los casos positivos del Covid 19 en el Atlantico. Mientras que el restante 97% corresponde a los recuperados.

## Dias de recuperacion

```python
plt.figure(figsize=(6,4))
plt.hist(df["Dias de recuperacion"], bins=50)
plt.xlabel("Días de recuperación")
plt.ylabel("Frecuencia")
plt.title("Histograma días de recuperación")
plt.show()


stats = df["Dias de recuperacion"].describe()

# Convertir a DataFrame y renombrar
tabla_dias_de_recuperacion = pd.DataFrame({
    "Estadístico": [
        "Número de casos",
        "Media",
        "Desviación estándar",
        "Mínimo",
        "Percentil 25 (Q1)",
        "Mediana (Q2)",
        "Percentil 75 (Q3)",
        "Máximo"
    ],
    "Valor": [
        int(stats["count"]),
        round(stats["mean"], 2),
        round(stats["std"], 2),
        int(stats["min"]),
        int(stats["25%"]),
        int(stats["50%"]),
        int(stats["75%"]),
        int(stats["max"])
    ]
})

tabla_dias_de_recuperacion
```

En este Grafico del histograma se puede analizar que existe una distribucion asimetrica con un gran sesgo hacia la derecha. En la tabla se puede observar que el 50% de los datos de las personas para recuperarse del covid 19 ronda entre los 14 y 21 dias. la media de dias de recuperacion es de 23 dias por lo que esta fuera de este rango intercuartilico debido a la gran dispersion de datos por la cola larga en la derecha, ya que existen pacientes con recuperacion de hasta 386 dias. En este caso la mejor medida tendencia central es la mediana con un valor de 15 dias al presentar mejor comportamiento tipico que la media
